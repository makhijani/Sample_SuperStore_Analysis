```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("SampleSuperstore.csv")
data
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
data.info()
```

```python
data.drop(columns="Postal Code",inplace=True)
data
```

```python
print(data["Country"].unique())
print(data["Region"].unique())
print(data["Category"].unique())
print(data["Segment"].unique())
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace = True)
```

```python
# statistical analysis of sales based on region
data.groupby("Region")["Sales"].sum().plot.bar()
```

**It means sales is highest in west region and lowest in south region.**

```python
# statistical analysis of profit based on region
data.groupby("Region")["Profit"].sum().plot.bar()
```

**West region has highest sale as well as highest profit. Central region have more sales than south region but south region have made more profit than central althought sales are less.**

```python
# statistical analysis of profit based on region using pie chart
data.groupby("Region")["Profit"].sum().plot.pie(autopct="%1.0f%%")
```

```python
#Statistical analysis of Sales based on Category
data.groupby("Category")["Sales"].sum().plot.pie(autopct="%1.0f%%")
```

**Technology has highest sale. Although difference in sales of all the categories are not much.**

```python
#Statistical analysis of Profit based on Category
data.groupby("Category")["Profit"].sum().plot.pie(autopct="%1.0f%%")
```

```python
#Statistical analysis of Sales based on Segment
data.groupby("Segment")["Sales"].sum().plot.pie(autopct="%1.0f%%")
```

```python
#Statistical analysis of Profit based on Segment
data.groupby("Segment")["Profit"].sum().plot.pie(autopct="%1.0f%%")
```

**Highest sales comes from Customer segment, also Customer segment gives highest profit**

**ALSO  Lowest sales comes from Home Office segment, also Home Office segment gives lowest profit**

```python
data["State"].unique()
```

```python
#Statistical analysis of Sales based on State
data.groupby("State")["Sales"].sum().plot.bar()
```

```python
#Statistical analysis of Profit based on State
data.groupby("State")["Profit"].sum().plot.bar()
```

```python
ps=data.groupby("Sub-Category")["Sales","Profit"].sum()
ps
```

```python
#Statistical analysis of Sales vs Profit in Sub-Categories
ps.plot.bar(color=['magenta','orange'],figsize=(16,9))
plt.title("Profit vs Sales in Sub-category")
plt.show()
```

**Copiers and Phones has highest sales, although Phones gives more profit then Copiers and Fasteners has lowest sales as well as profit**

**ALSO Bookcases and Tables are in loss**

```python
#Statistical analysis of Sales based on Shipping mode
data.groupby("Ship Mode")["Sales"].sum().plot.bar()
```

**Standard Class has highest sales and Same Day has lowest it means that the customers are preferring shipping by standard class**
